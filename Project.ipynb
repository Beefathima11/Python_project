```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\beefa\\OneDrive\\Desktop\\Project.csv")
```

```python
# Preprocessing: Replace 'height' column with random values between 150-180 cm
df['Height'] = np.random.randint(150, 181, df.shape[0])
```

```python
# 1. Employee Distribution Across Teams
distribution = df['Team'].value_counts()
percentage_distribution = (distribution / len(df)) * 100

plt.figure(figsize=(10,5))
sns.barplot(x=distribution.index, y=distribution.values, palette='viridis')
plt.xlabel("Teams")
plt.ylabel("Number of Employees")
plt.title("Employee Distribution Across Teams")
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Segregation by Position
position_count = df['Position'].value_counts()
plt.figure(figsize=(10,5))
sns.barplot(x=position_count.index, y=position_count.values, palette='coolwarm')
plt.xlabel("Position")
plt.ylabel("Count")
plt.title("Employee Count by Position")
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Predominant Age Group
bins = [20, 30, 40, 50, 60]
labels = ['20-30', '31-40', '41-50', '51-60']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
age_group_count = df['Age Group'].value_counts()
plt.figure(figsize=(7,5))
sns.barplot(x=age_group_count.index, y=age_group_count.values, palette='magma')
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.title("Distribution of Employees by Age Group")
plt.show()
```

```python
# 4. Highest Salary Expenditure
team_salary = df.groupby('Team')['Salary'].sum().sort_values(ascending=False)
position_salary = df.groupby('Position')['Salary'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
sns.barplot(x=team_salary.index, y=team_salary.values, palette='Blues')
plt.xlabel("Teams")
plt.ylabel("Total Salary Expenditure")
plt.title("Total Salary by Team")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10,5))
sns.barplot(x=position_salary.index, y=position_salary.values, palette='Reds')
plt.xlabel("Positions")
plt.ylabel("Total Salary Expenditure")
plt.title("Total Salary by Position")
plt.xticks(rotation=45)
plt.show()
```

```python
# 5. Correlation Between Age and Salary
plt.figure(figsize=(7,5))
sns.scatterplot(x=df['Age'], y=df['Salary'], alpha=0.6)
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs. Salary Correlation")
plt.show()

correlation = df[['Age', 'Salary']].corr().iloc[0,1]
print(f"Correlation between Age and Salary: {correlation:.2f}")
```
